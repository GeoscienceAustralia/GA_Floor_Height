```python

# Paths to input folders and GeoJSON file
geojson_file = r'C:\Users\lliu\FrontierSI\Projects - 127 Residential Dwelling Floor Height\4 Executing\Data Exploration\GSV\Wagga\Panos\streetlevel\Final_Wagga_training_samples_meta_split.geojson'
image_folder = r"C:\Users\lliu\FrontierSI\Projects - 127 Residential Dwelling Floor Height\4 Executing\Data Exploration\GSV\Wagga\YOLO_test\finetune\all\images"
label_folder = r"C:\Users\lliu\FrontierSI\Projects - 127 Residential Dwelling Floor Height\4 Executing\Data Exploration\GSV\Wagga\YOLO_test\finetune\all\labels"
output_folder = r"C:\Users\lliu\FrontierSI\Projects - 127 Residential Dwelling Floor Height\4 Executing\Data Exploration\GSV\Wagga\YOLO_test\finetune"  # Root folder for split directories
```

```python

# Read the GeoJSON file
geo_df = gpd.read_file(geojson_file)

# Ensure the output folder structure exists
os.makedirs(output_folder, exist_ok=True)

# Iterate through the GeoDataFrame rows
for _, row in geo_df.iterrows():
    pano_id = row['pano_id']  # Extract pano_id
    split = row['split']      # Extract split (e.g., train, validation)
    
    # Skip rows with empty pano_id
    if not pano_id or pd.isna(pano_id):
        print("Skipping row with empty pano_id")
        continue
    # Define source and destination paths for images and labels
    # image_src = os.path.join(image_folder, f"{pano_id}.png")  # Update extension if needed
    # label_src = os.path.join(label_folder, f"{pano_id}.png")  # Update extension if needed
    image_src = os.path.join(image_folder, f"{pano_id}.jpg")  # Update extension if needed
    label_src = os.path.join(label_folder, f"{pano_id}.txt")  # Update extension if needed
    
    split_image_folder = os.path.join(output_folder, split, "images")
    split_label_folder = os.path.join(output_folder, split, "labels")
    
    # Create split-specific folders if they don't exist
    os.makedirs(split_image_folder, exist_ok=True)
    os.makedirs(split_label_folder, exist_ok=True)
    
    # Define destination paths
    # image_dst = os.path.join(split_image_folder, f"{pano_id}.png")
    # label_dst = os.path.join(split_label_folder, f"{pano_id}.png")
    image_dst = os.path.join(split_image_folder, f"{pano_id}.jpg")
    label_dst = os.path.join(split_label_folder, f"{pano_id}.txt")
    
    # Copy files to the respective split folders
    if os.path.exists(image_src):
        shutil.copy(image_src, image_dst)
    else:
        print(f"Image file missing: {image_src}")
    
    if os.path.exists(label_src):
        shutil.copy(label_src, label_dst)
    else:
        print(f"Label file missing: {label_src}")

print(f"Files have been organized into the '{output_folder}' directory based on the 'split' attribute.")
```
