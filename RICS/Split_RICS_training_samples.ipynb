```python
import os
import shutil
import geopandas as gpd
import pandas as pd
```

```python

# Paths to input folders and GeoJSON file
geojson_file = r'C:\Users\lliu\Desktop\FrontierSI\projects\GA_floor_height\GA-floor-height\output\Wagga\RICS\RICS_wagga_training_locations\Final_Wagga_training_samples_RICS_split.geojson'
image_folder = r"D:\Wagga\RICS\annotations_converted_merged\images"
label_folder = r"D:\Wagga\RICS\annotations_converted_merged\labels"
output_folder = r"D:\Wagga\RICS\annotations_converted_merged"  # Root folder for split directories
```

```python

# Read the GeoJSON file
geo_df = gpd.read_file(geojson_file)

# Ensure the output folder structure exists
os.makedirs(output_folder, exist_ok=True)

# Iterate through the GeoDataFrame rows
for _, row in geo_df.iterrows():
    img_id = row['RICS_img']  # Extract image id
    split = row['split']      # Extract split (e.g., train, validation)
    
    # Skip rows with empty pano_id
    if not img_id or pd.isna(img_id):
        print("Skipping row with empty pano_id")
        continue
    # Define source and destination paths for images and labels
    image_src = os.path.join(image_folder, img_id).replace('.jpg','.png')  # Update extension if needed
    label_src = os.path.join(label_folder, img_id).replace('.jpg','.png')  # Update extension if needed
    
    split_image_folder = os.path.join(output_folder, split, "images")
    split_label_folder = os.path.join(output_folder, split, "labels")
    
    # Create split-specific folders if they don't exist
    os.makedirs(split_image_folder, exist_ok=True)
    os.makedirs(split_label_folder, exist_ok=True)
    
    # Define destination paths
    image_dst = os.path.join(split_image_folder, img_id).replace('.jpg','.png')
    label_dst = os.path.join(split_label_folder, img_id).replace('.jpg','.png')
    
    # Copy files to the respective split folders
    if os.path.exists(image_src):
        shutil.copy(image_src, image_dst)
    else:
        print(f"Image file missing: {image_src}")
    
    if os.path.exists(label_src):
        shutil.copy(label_src, label_dst)
    else:
        print(f"Label file missing: {label_src}")

print(f"Files have been organized into the '{output_folder}' directory based on the 'split' attribute.")
```
